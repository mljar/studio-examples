```python
import pandas as pd
# load example dataset
df = pd.read_csv("https://raw.githubusercontent.com/pplonski/datasets-for-start/master/adult/data.csv", skipinitialspace=True)
# display DataFrame shape
print(f"Loaded data shape {df.shape}")
# display first rows
df.head()
```

```python
from sklearn.model_selection import train_test_split
# split data
train, test = train_test_split(df, train_size=0.75, shuffle=True, random_state=42)
# display data shapes
print(f"All data shape {df.shape}")
print(f"Train shape {train.shape}")
print(f"Test shape {test.shape}")
```

```python
import pandas as pd
# create X columns list and set y column
x_cols = ["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", "hours-per-week", "native-country"]
y_col = "income"
# set input matrix
X = train[x_cols]
# set target vector
y = train[y_col]
# display data shapes
print(f"X shape is {X.shape}")
print(f"y shape is {y.shape}")
```

```python
from supervised import AutoML
# create automl object
automl = AutoML(total_time_limit=300, mode="Explain")
# train automl
automl.fit(X, y)
```

```python
# display AutoML report
automl.report()
```

```python
import pandas as pd
# create X columns list and set y column
x_cols = ["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", "hours-per-week", "native-country"]
y_col = "income"
# set input matrix
X_test = test[x_cols]
# set target vector
y_test = test[y_col]
# display data shapes
print(f"X_test shape is {X_test.shape}")
print(f"y_test shape is {y_test.shape}")
```

```python
# predict with AutoML
predictions = automl.predict(X_test)
# predicted values
print(predictions)
```

```python
from sklearn.metrics import accuracy_score
# compute Accuracy
metric_accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {metric_accuracy}")
```
